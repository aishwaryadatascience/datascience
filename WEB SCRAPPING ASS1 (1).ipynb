```python
#Answer 1

import requests
from bs4 import BeautifulSoup
import pandas as pd

#Send an HTTP GET request to wikipedia.org
url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url)

#Check if the request was successful
if response.status_code ==200:
    #Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    #Find all header tags (h1, h2, h3, h4, h5, h6)
    header_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
    
    #Extract the text from the header tags and store them in a list 
    header_texts = [tag.text.strip() for tag in header_tags]
    
    #Create a DataFrame from the extracted header texts
    df = pd.DataFrame({'Header':header_texts})
    
    #Display the DataFrame
    print(df)
    
else:
    print ("Failed to retrive the web page. Status code:", response.status_code)
    
```

```python
#Answer3

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape and create a DataFrame for rankings
def scrape_icc_rankings(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find the relevant table
        table = soup.find('table', class_='table')
        
        # Initialize lists to store data
        names = []
        matches = []
        points = []
        ratings = []

        # Find and extract data from the table
        for row in table.find_all('tr')[1:11]:  # Extract top 10 rows
            columns = row.find_all('td')
            if len(columns) == 5:  # Ensure the row has the expected number of columns
                name = columns[1].text.strip()
                match = columns[2].text.strip()
                point = columns[3].text.strip()
                rating = columns[4].text.strip()
                names.append(name)
                matches.append(match)
                points.append(point)
                ratings.append(rating)
        
        # Create a DataFrame
        df = pd.DataFrame({'Team': names, 'Matches': matches, 'Points': points, 'Rating': ratings})
        
        return df
    else:
        print(f"Failed to retrieve the web page. Status code: {response.status_code}")
        return None

# URLs for different rankings
odi_teams_url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
odi_batsmen_url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
odi_bowlers_url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"

# Scrape and create DataFrames
odi_teams_df = scrape_icc_rankings(odi_teams_url)
odi_batsmen_df = scrape_icc_rankings(odi_batsmen_url)
odi_bowlers_df = scrape_icc_rankings(odi_bowlers_url)

# Display DataFrames
if odi_teams_df is not None:
    print("Top 10 ODI Teams:")
    print(odi_teams_df)

if odi_batsmen_df is not None:
    print("\nTop 10 ODI Batsmen:")
    print(odi_batsmen_df)

if odi_bowlers_df is not None:
    print("\nTop 10 ODI Bowlers:")
    print(odi_bowlers_df)
```
